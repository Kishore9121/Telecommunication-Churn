```python
## outliers removal
```

```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data_clean = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return data_clean

# Separate continuous variables (excluding 'churn' column)
continuous_columns = df.select_dtypes(include=[np.number]).columns

# Remove outliers for each continuous variable
for column in continuous_columns:
    df = remove_outliers_iqr(df, column)

# Resulting dataset with outliers removed
print(df)
```

```python
df.boxplot(figsize=(20,10))
plt.title("boxplot after removing outliers")
```
